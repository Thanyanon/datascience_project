```python
# create dictionary to store a data from API

anime_dict = {
    'id' : [],
    'url' : [],
    'img' : [],
    'title' : [],
    'title_english' : [],
    'type' : [],
    'episode' : [],
    'status' : [],
    'aired' : [],
    'duration' : [],
    'rating' : [],
    'score' : [],
    'score_by' : [],
    'member' : [],
    'favorite' : [],
    'studio' : [],
    'genre' : []
    }
```

```python
# dictionary keys
anime_dict.keys()
```

```python
# getting anime data via API

data = list()
id = []
url = []
img = []
title = []
title_english = []
types = []
episode = []
status = []
aired = []
duration = []
rating = []
score = []
scored_by = []
rank = []
member = []
favorite = []
studio = []
producer = []
genre = []

for i in range(1, page_num+1):
    page_url = 'https://api.jikan.moe/v4/anime?page={}'
    page_url = page_url.format(i)
    response = requests.get(page_url)
    if(response.status_code == 200) :
        result = response.json()
        for i in range(len(result['data'])) :
            data = result.get('data')[i]
            try :
                id.append(data.get('mal_id'))
                url.append(data.get('url'))
                img.append(data.get('images').get('jpg').get('image_url'))
                title.append(data.get('title')),
                title_english.append(data.get('title_english'))
                types.append(data.get('type'))
                episode.append(data.get('episodes'))
                status.append(data.get('status'))
                aired.append(data.get('aired').get('string'))
                duration.append(data.get('duration'))
                rating.append(data.get('rating'))
                score.append(data.get('score'))
                scored_by.append(data.get('scored_by'))
                member.append(data.get('members'))
                favorite.append(data.get('favorites'))
                studio.append(data.get('studios')[0].get('name'))
                genre.append([data.get('genres')[i].get('name') for i in range(len(data.get('genres')))])
            except IndexError:
                studio.append('N/A')
                genre.append([data.get('genres')[i].get('name') for i in range(len(data.get('genres')))])
        else: pass
    time.sleep(0.5)
```

```python
# store data to dictionary

anime_dict = {
    'id' : id,
    'url' : url,
    'img' : img,
    'title' : title,
    'title_english' : title_english,
    'type' : types,
    'episode' : episode,
    'status' : status,
    'aired' : aired,
    'duration' : duration,
    'rating' : rating,
    'score' : score,
    'score_by' : scored_by,
    'member' : member,
    'favorite' : favorite,
    'studio' : studio,
    'genre' : genre
    }
```

```python
# Number of data

length_dict = {key: len(value) for key, value in anime_dict.items()}
length_dict
```

```python
# Convert dict to dataframe

anime_df = pd.DataFrame.from_dict(anime_dict)
```

```python
# write to csv

anime_df.to_csv('myanimelist.csv', index = False)
```
