## List of player link

- create a list of all proplayer link in [prosetting.net](https://prosettings.net/players/) website for scraping.

```python
# total page of player

url = "https://prosettings.net/players/"

req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')
page = soup.find_all('a', {'class' : 'page-numbers'})
page = [page.text for page in page]
page = page[-2]
print(f'total page of player = {page} pages')
```

```python
# create a list contained all of player url link

url = "https://prosettings.net/players/page/"

player_link = []

for page in range(1, int(page) + 1) :
    req = requests.get(url + str(page) + '/') # select multiple page
    soup = BeautifulSoup(req.text, 'html.parser')
    link = soup.find('div', {'class', 'cards-grid--list js-players-list'}) # get a table of player link
    players = link.find_all('a') # find an url of player link
    for i in range(len(players)) :
        link = players[i]['href']
        player_link.append(link)
        print(link)
```

```python
print(f'total player in prosetting website = {len(player_link)} players')
```

## Bio data

- scraping pro players bio data.
- get player bio data
- bio data contain :
    1. ign or in game name
    2. name and surname
    3. date of birth
    4. country
    5. team
    6. game

```python
# scraping a pro player bio data

player_bio = []

# get the bio data from all player in the web
for link in player_link:
    req = requests.get(link)
    soup = BeautifulSoup(req.text, 'html.parser')
    
    # soup bio data from web
    bio = soup.find('div', {"class": "center-container"})
    
    # get ingame name, name, birthdate, country, team and game from bio data
    ign = bio.find('h1').text # in game name
    name = bio.find('tr', {'class': 'format-text field-name'}).find('td').text
    birth = bio.find('tr', {'class': 'format-date_picker field-birthday'}).find('td').text
    country = bio.find('tr', {'class': 'format-country field-country'}).find('span').text
    team = bio.find('tr', {'class': 'format-text field-team'}).find('td').text
    game = bio.find('div', {'class': 'games'}).find('h2').text.split(' ')[0]
    print(f'{ign} {name} {birth} {country} {team} {game}')
    player_bio.append([ign, name, birth, country, team, game])
```

```python
# total bio player data

print(f'total player bio data = {len(player_bio)} players')
player_bio[:3]
```

```python
# create a dataframe of player bio data

player_bio_df = pd.DataFrame(player_bio ,columns = ['ign', 'name', 'birth', 'country', 'team', 'game'])

player_bio_df
```

```python
# create an ages columns

# Define a function to calculate the age based on the birth date
def calculate_age(row):
    if pd.isnull(row['birth']):
        return float('nan')
    birth_date = row['birth']
    today = datetime.now()
    age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return age

# Apply the function to the DataFrame and assign the result to a new column
player_bio_df['age'] = player_bio_df.apply(calculate_age, axis=1)

player_bio_df
```

```python
# summary statistics of age column

player_bio_df['age'].describe()
```

```python
# create age_range columns

## define age range function

def cal_age_range(row):
    if pd.isnull(row['age']):
        return float('nan')
    age = row['age']
    if age >= 0 and age <= 9:
        return 'less than 10'
    elif age >= 10 and age <= 14:
        return '10-14'
    elif age >= 15 and age <= 19:
        return '15-19'
    elif age >= 20 and age <= 24:
        return '20-24'
    elif age >= 25 and age <= 29:
        return '25-29'
    elif age >= 30 and age <= 34:
        return '30-34'
    elif age >= 35 and age <= 39:
        return '35-39'
    elif age >= 40 and age <= 44:
        return '40-44'
    else:
        return 'more than 44'
    
player_bio_df['age_range'] = player_bio_df.apply(cal_age_range, axis=1)

player_bio_df['age_range'].unique()
```

```python
# replace country that cannot search with search_fuzzy() function from pycountry

player_bio_df['country'] = player_bio_df['country'].replace(['Bosnia & Herzegovina', 'Hong Kong SAR China'], 
                                                            ['Bosnia and Herzegovina', 'Hong Kong'])

player_bio_df['country'].unique()
```

```python
# create a region column from country

import pycountry
import pycountry_convert as pc

# Define a function to get the region for a given country name
def get_region_from_country(country_name):
    # Use the search_fuzzy() function to search for a country by name
    # The search is fuzzy, so it will return the best match even if the country name is not an exact match
    country_code = pycountry.countries.search_fuzzy(country_name)[0].alpha_2

    # use the country_alpha2_to_continent_code() function to convert the country code to a continent code
    continent_code = pc.country_alpha2_to_continent_code(country_code)
    
    # Return the continent code
    return continent_code

# Use the apply() function to apply the get_region_from_country() function to each value in the "Country" column
player_bio_df['continent'] = player_bio_df['country'].apply(get_region_from_country)

player_bio_df
```

```python
# create a correcting factor columns for sensitivity

ratio = {
    'CS:GO' : 1,
    'VALORANT' : 3.18,
    'Overwatch' : 0.3,
    'Fortnite' : 0.25,
    'Apex' : 1,
    'PUBG' : 0.24,
    'Rainbow' : 0.26,
    'Call of duty' : 0.3
}

player_bio_df['edpi_ratio'] = player_bio_df['game'].map(lambda x: ratio.get(x, 0))

player_bio_df
```

```python
# rearange the columns

player_bio_df = player_bio_df.reindex(columns=['ign', 'name', 'age','age_range', 'birth', 'country', 'continent', 'team', 'game', 'edpi_ratio'])

player_bio_df
```

## Mouse

- scrap a pro players mouse setting.
- setting contain:
    1. mouse name
    2. sensitivity = speed of in game mouse
    3. polling rate = a frequency of mouse sensor
    4. dpi = speed of mouse or dot per inch
    5. edpi = dpi * sensitivity which mean a **true mouse speed**
- setting data from 868 players.
- setting data are incomplete, some player have a missing value and some don't have a setting due to using controller instead of mouse.

**Note:**
    
    -  scrap   try : except  'N/A'  scrap 
    -   continue  except  scrap link 
    - scaping ign  primary key

```python
# scraping a pro player mouse setting data

player_mouse = []

for link in player_link:
    req = requests.get(link)
    soup = BeautifulSoup(req.text, 'html.parser')
    
    # get a ign from bio data
    bio = soup.find('div', {'class': 'center-container'})
    ign = bio.find('h1').text # in game name
    
    # get a mouse setting from mouse setting section
    try:
        mouse = bio.find('section', {'class' : 'settings-group section--mouse'})
    except AttributeError:
        continue
    try:
        mouse_name = mouse.find('h4').text # mouse name
    except AttributeError:
        mouse_name = np.nan
    try:
        setting = mouse.find('table')
    except AttributeError:
        continue
    try:
        sense = setting.find('tr', {'data-field' : 'sensitivity'}).text.split('\n')[2] # sensivity
    except AttributeError:
        sense = np.nan
    try:
        hz = setting.find('tr', {'data-field' : 'hz'}).text.split('\n')[2] # polling rate
    except AttributeError:
        hz = np.nan
    try:
        dpi = setting.find('tr', {'data-field' : 'dpi'}).text.split('\n')[2] # dpi
    except AttributeError:
        dpi = np.nan
    try:
        edpi = setting.find('tr', {'data-field' : 'edpi'}).text.split('\n')[2] # edpi
    except AttributeError:
        edpi = np.nan
    print(f'{ign}, {mouse_name}, {sense}, {hz}, {dpi}, {edpi}')
    player_mouse.append([ign, mouse_name, sense, hz, dpi, edpi])
    
```

```python
# total player mouse setting data

print(f'total player mouse setting data = {len(player_mouse)} players')
player_mouse[:5]
```

```python
# create a dataframe of mouse setting

player_mouse_df = pd.DataFrame(player_mouse, columns = ['ign', 'mouse', 'sense', 'hz', 'dpi', 'edpi'])

player_mouse_df
```

```python
# drop missing values

player_mouse_df = player_mouse_df.dropna().reset_index(drop=True).copy()

print(player_mouse_df.isna().sum())
print(player_mouse_df.shape)
```

```python
# create a mouse brand columns from mouse columns
import re

# Use the apply method to apply the re.search function to each row in the 'sentence' column
def extract_first_word(sentence):
    try:
        match = re.search(r"\S+\s", sentence)
        if match:
            return match.group().strip()
        return None
    except TypeError:
        return np.nan

player_mouse_df['mouse_brand'] = player_mouse_df['mouse'].apply(extract_first_word)

player_mouse_df['mouse_brand'].unique()
```

```python
# rearange the columns

player_mouse_df = player_mouse_df.reindex(columns=['ign', 'mouse', 'mouse_brand', 'sense', 'hz', 'dpi', 'edpi'])

player_mouse_df
```

## Gear

- scrap a pro players gear list.
- gear list contain:
    1. mouse
    2. mouse pad
    3. keyboard
    4. monitor
    5. headset
    6. earphones
    7. controller
- gear list from 862 players.
- gear list data are incomplete, some player have a missing value of gear list

**Note:**

    -    monitor  scrap  gear  ign  label  value  gear   ign  tuple   dataframe  key:value pair tuple  transform dataframe  long  wide format

```python
# scraping the player gear

player_gear = ()

for link in player_link:
    req = requests.get(link)
    soup = BeautifulSoup(req.text, 'html.parser')
    
    # get a ign from bio data
    bio = soup.find('div', {'class': 'center-container'})
    ign = bio.find('h1').text # in game name
    
    # get a game resolution
    resolution = bio.find('tr', {'data-field' : 'resolution'}).text.split('\n')[2] # resolution
    
    # get a gear list
    gear = soup.find('section', {'id' : 'gear'})
    try:
        gear_list = [gear.text for gear in gear.find_all('div', {'class' : 'cta-box_status-tag'})]
        print(gear_list)
        gear_name = [gear.text for gear in gear.find_all('h4')]
        print(gear_name)
        player_name = []
        [player_name.append(ign) for i in range(len(gear_list))]
        print(player_name)
        player_gear += (tuple(zip(gear_list, gear_name, player_name)))
    except AttributeError:
        continue
```

```python
# replace headset NaN value with earphones

def fill_na(df, column, other_column):
    def fill_na_row(row):
        if pd.isnull(row[column]):
            return row[other_column]
        return row[column]

    df[column] = df.apply(fill_na_row, axis=1)

fill_na(player_gear_df, 'headset', 'earphones')
```
